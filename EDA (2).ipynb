### Q8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
# Displaying the heatmap for the required columns
sns.heatmap(df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb',
       'prodvalue']].corr(), annot = True, )
```

### Inferences : 
1. Numcol and totalprod have correlation of 0.95 which represent that there is very strong high positive correlation between them.
2. Numcol and stocks have correlation of 0.83 which represent that there is high positive correlation between them.
